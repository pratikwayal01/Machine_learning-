```python
from sklearn.model_selection import train_test_split

# Extract the specified columns
selected_columns = ['Age', 'Sex', 'ChestPain', 'RestBP', 'Chol']
df_selected = df[selected_columns]

# Split the data into training and testing sets
train_data, test_data = train_test_split(df_selected, test_size=0.25, random_state=42)

print("Training data shape:", train_data.shape)
print("Testing data shape:", test_data.shape)
```

```python
#  Through the diagnosis test, I predicted 100 reports as COVID positive, but only 45 of those were actually positive. Total 50 people in my sample were actually COVID positive. Create a confusion matrix based on the above data and find:
#   I. Accuracy âœ…
#   II. Precision âœ¨
#   III. Recall ðŸ“¢
#   IV. F-1 score ðŸ“ˆ

# True Positives (TP): Correctly predicted COVID positive
TP = 45

# False Positives (FP): Predicted COVID positive but actually negative
FP = 100 - TP

# True Negatives (TN): Correctly predicted COVID negative (not needed for calculation)
# TN = ...

# False Negatives (FN): Predicted COVID negative but actually positive
FN = 50 - TP

# I. Accuracy
accuracy = (TP) / (TP + FP + FN)
print("Accuracy:", accuracy)

# II. Precision
precision = TP / (TP + FP)
print("Precision:", precision)

# III. Recall
recall = TP / (TP + FN)
print("Recall:", recall)

# IV. F-1 Score
f1_score = 2 * (precision * recall) / (precision + recall)
print("F1 Score:", f1_score)
```
